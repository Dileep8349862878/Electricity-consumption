```python
def create_features(df):

    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    df['weekofyear'] = df.index.isocalendar().week
    return df

df = create_features(df)
```

```python
train = create_features(train)
test = create_features(test)
```

```python
FEATURES = ['dayofyear', 'hour', 'dayofweek', 'quarter', 'month', 'year']
TARGET = 'PJME_MW'

X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
test['prediction'] = reg.predict(X_test)
df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)
ax = df[['PJME_MW']].plot(figsize=(15, 5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f'RMSE Score on Test set: {score:0.2f}')
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
import pmdarima as pm

model = pm.auto_arima(y_train, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)

print(model.summary())
```

```python
train_size = int(0.85 * len(df))
test_size = len(df) - train_size

train, test = df.iloc[:train_size], df.iloc[train_size:]

# Features and Target
FEATURES = ['dayofyear', 'hour', 'dayofweek', 'quarter', 'month', 'year']
TARGET = 'PJME_MW'

X_train = train[FEATURES]
y_train = train[TARGET]
X_test = test[FEATURES]
y_test = test[TARGET]

# ARIMA Model
arima_model = ARIMA(y_train, order=(3, 0, 0))  # Adjust order as necessary

# Fit the ARIMA model
model_fit = arima_model.fit()

# ARIMA predictions (forecast for test size)
y_pred_arima = model_fit.forecast(steps=test_size)

# Add ARIMA predictions to the test set
test['arima_prediction'] = y_pred_arima.values

# Assuming you have a regression model (e.g., XGBoost or RandomForest) already fitted as `reg`
# Make regression predictions
test['regression_prediction'] = reg.predict(X_test)

# Merge predictions back into the original df for plotting
df = df.merge(test[['arima_prediction', 'regression_prediction']], how='left', left_index=True, right_index=True)

# Plot actual values and predictions
ax = df[['PJME_MW']].plot(figsize=(15, 5), label='Truth Data')
df['arima_prediction'].plot(ax=ax, style='--', color='red', label='ARIMA Predictions')
df['regression_prediction'].plot(ax=ax, style='.', color='blue', label='Regression Predictions')

# Plot settings
plt.legend(['Truth Data', 'ARIMA Predictions', 'Regression Predictions'])
ax.set_title('Actual Data and Predictions')
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f'RMSE Score on Test set: {score:0.2f}')
```
